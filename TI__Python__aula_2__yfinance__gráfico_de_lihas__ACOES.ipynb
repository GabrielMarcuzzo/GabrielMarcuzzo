```python
# Importe as bibliotecas necessárias
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Defina os três tickers
tickers = ['USIM5.SA', 'SAPR4.SA', 'AURE3.SA']  # Você pode trocar por outros tickers
```

```python
# Baixe os dados históricos dos últimos 2 anos
dados = yf.download(tickers, start="2023-01-01")['Close']
```

```python
# Plotando o gráfico de linhas
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(dados[ticker], label=ticker)

plt.title('Preço de Fechamento das Ações')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
