```python
#  PASSO 2: Importar a biblioteca plotly

import plotly.express as px
import pandas as pd
```

```python
#  PASSO 3: Definir parâmetros do projeto


data = {
    "Atividade": [
        "Atividade_1",
        "Atividade_2",
        "Atividade_3",
        "Atividade_4",
        "Atividade_5"
    ],
    "Início": [
        "2026-03-01",
        "2026-03-04",
        "2026-03-10",
        "2026-03-20",
        "2026-03-25"
    ],
    "Fim": [
        "2026-03-04",
        "2026-03-09",
        "2026-03-19",
        "2026-03-24",
        "2026-03-28"
    ]
      }
```

```python
# PASSO 4: Criar o DataFrame

df = pd.DataFrame(data)
```

```python
# PASSO 5: Criar o Gráfico de Gantt
fig = px.timeline(df, x_start="Início", x_end="Fim", y="Atividade")
fig.update_layout(title="Gráfico de Gantt", xaxis_title="Datas", yaxis_title="Atividades")

# Plotar o gráfico
fig.show()
```
