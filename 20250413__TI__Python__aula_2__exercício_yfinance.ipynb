```python
import yfinance as yf
import matplotlib.pyplot as plt

# Defina os três tickers (ações da B3 devem terminar com .SA)
tickers = ['PETR4.SA', 'VALE3.SA', 'ITUB4.SA']  # Você pode trocar por outros tickers

# Baixe os dados históricos dos últimos 2 anos
dados = yf.download(tickers, start="2023-01-01")['Close']

# Plotando o gráfico de linhas
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(dados[ticker], label=ticker)

plt.title('Preço de Fechamento das Ações')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
