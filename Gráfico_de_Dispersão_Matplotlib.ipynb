```python
# Gráfico de Dispersão criado com a biblioteca Matplotlib

# Importar a biblioteca
import matplotlib.pyplot as plt

# Suas listas de exemplo
metragem = [120, 230, 80, 90, 140, 160, 300, 60, 110]
valor = [700, 1200, 550, 500, 800, 750, 400, 400, 350]

# Criando o gráfico de dispersão
plt.scatter(metragem, valor, color="blue", alpha=0.7, edgecolors="k")

# Adicionando título e rótulos
plt.title("Gráfico de Dispersão - Metragem vs Valores de Imóveis")
plt.xlabel("Metragem")
plt.ylabel("Valor do Imóvel")
plt.grid(True)

# Exibindo
plt.show()
```
